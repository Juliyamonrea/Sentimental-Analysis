```python
import pandas as pd
import matplotlib.pyplot as plt

# read the CSV files into Pandas dataframes
df1 = pd.read_csv("ola_keyword.csv")
df2 = pd.read_csv("ather_keyword.csv")

# create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 5))

# create the first bar chart in the left subplot
ax1.barh(df1["keywords"], df1["count"], color='green')
ax1.set_xlabel("Frequency")
ax1.set_ylabel("Keywords")
ax1.set_title("Ola S1 sentimental analysis", fontweight="bold")
ax1.set_xlim(0, max(df1["count"]) + 5)
ax1.set_yticks(df1.index)
ax1.set_yticklabels(df1["keywords"], rotation=0)
ax1.spines["right"].set_visible(False)

# create the second bar chart in the right subplot
ax2.barh(df2["keywords"], df2["count"], color='black')
ax2.set_xlabel("Frequency")
ax2.set_ylabel("Keywords")
ax2.set_title("Ather 450x Gen3 sentimental analysis", fontweight="bold")
ax2.set_xlim(0, max(df2["count"]) + 5)
ax2.set_yticks(df2.index)
ax2.set_yticklabels(df2["keywords"], rotation=0)
ax2.yaxis.tick_right()
ax2.spines["left"].set_visible(False)

# adjust the spacing between the subplots
plt.subplots_adjust(wspace=0.05)

# show the figure
plt.show()
```
