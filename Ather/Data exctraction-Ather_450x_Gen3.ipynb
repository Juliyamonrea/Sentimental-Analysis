```python
api_key = 'AIzaSyBhrhxPGC6zJ13vpZUg6aK00kGxumr5Y7Q'#Add YouTube V3 API Key
video_ids = ['bRyL8nqCXiA',
'ntylQp76Deg',
'i3ijIQyISlM',
'bHC27Z92n_o',
'VtdshvBbjVA',
'7yE4Cl_LxuE',
'pX-FGzaoPPA',
'rOLNO4HfRqc',
'WcO1XJuan2M',
'wx_Of1Nejuk',
'UYsIRezBrA8',
'A0Cy4tONjPs',
'5MSC3k6YXn8',
'Mz5SxPDaj2U',
'F4Jug0iyt2E',
'n5rfvWSF5u8',
'a3RqOJ1AMco',
'MIOaoukFQWg',
'Ed3Lv2wGeU8',
'b2t0EprvT3A',
'mMxSstgdha4',
'mtbtROH6pi0',
'8XpUFCQaox8',
'CetyA8o9EPo',
'qVUKMns4_Rg',
'HLz-akGoHMY',
'P5k8uve04xc',
'N9Z2bysFCSg',
'58e_-XGHG74',
'uWNkRlTX_Ug',
'ybByG4nBuOc',
'r1bTACGfJ7s',
'-Td16gcNXc4',
'FzQlybnOGGI',
'TD_4RZRMKVY',
'6f0qRjEpsTY',
'hhlRfWiAewg',
'wRsGjVzza7c',
'mjQ5mWVvJaI',
'u4Wd5xISt_s',
'OFMxNe6pRp0',
'V9TDytIqZ9A',
'tWpg1pm8Jmw',
'j2G2BrYHaDo',
'rbLvzJ3tjvo',
'x10H492GeQw',
'GcVldPA0tkM',
'e6CeCqNylL4',
'58NG0Cgmbio',
'ryJKme9uQ9M'

]
def video_comments(video_id):
    # empty list for storing comments and replies
    comments = []

    # creating youtube resource object
    youtube = build('youtube', 'v3', developerKey=api_key)

    # retrieve youtube video results
    video_response=youtube.commentThreads().list(
        part='snippet,replies',
        videoId=video_id,
        textFormat="plainText"
    ).execute()

    # get total number of comments
    total_comments = video_response['pageInfo']['totalResults']

    # iterate video response
    while video_response:
        for item in video_response['items']:
            comment = item['snippet']['topLevelComment']['snippet']['textDisplay']
            # counting number of reply of comment
            replycount = item['snippet']['totalReplyCount']
            if replycount>0:
                for reply in item['replies']['comments']:
                    reply = reply['snippet']['textDisplay']
                    comments.append([reply])
            comments.append([comment])
            
        # check if there are more comments
        if 'nextPageToken' in video_response:
            video_response = youtube.commentThreads().list(
                part='snippet,replies',
                videoId=video_id,
                textFormat="plainText",
                pageToken=video_response['nextPageToken']
            ).execute()
        else:
            break

    return comments, total_comments

# create list to store all results
all_results = []

# iterate through each video ID and call the video_comments function
for video_id in video_ids:
    comments, total_comments = video_comments(video_id)
    all_results.append([video_id, total_comments, comments])

# write results to csv
with open('comments.csv', mode='w', encoding='utf-8', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Video ID', 'Total Comments', 'Comments'])
    for result in all_results:
        writer.writerow(result)
```
